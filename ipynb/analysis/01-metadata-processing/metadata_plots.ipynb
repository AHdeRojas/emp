```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_context("paper")
```

```python
colors = ['#1f78b4',
          '#33a02c',
          '#e31a1c',
          '#ff7f00',
          '#6a3d9a',
          '#a6cee3',
          '#b2df8a',
          '#fb9a99',
          '#fdbf6f',
          '#cab2d6',
          '#000000']
sns.set_palette(colors)
```

```python
path_map = '~/emp/metadata-refine/emp_qiime_mapping_refined_20160812_qc.tsv'
```

```python
df = pd.read_csv(path_map, sep='\t', index_col=0)
```

```python
df_temperature = df[df.temperature_deg_c.isnull() == False]
df_salinity    = df[df.salinity_psu.isnull() == False]
df_ph          = df[df.ph.isnull() == False]
df_oxygen      = df[df.oxygen_mg_per_l.isnull() == False]
```

```python
param = 'temperature_deg_c'
plt.figure(figsize=(30, 5))
sns.boxplot(x='study_id', y=param, hue='empo_3', data=df_temperature, linewidth=0.2)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
param = 'ph'
plt.figure(figsize=(30, 5))
sns.boxplot(x='study_id', y=param, hue='empo_3', data=df_ph, linewidth=0.2)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
param = 'salinity_psu'
plt.figure(figsize=(30, 5))
sns.boxplot(x='study_id', y=param, hue='empo_3', data=df_salinity, linewidth=0.2)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('/Users/luke/emp/plots/boxplot_%s_by_study.pdf' % param)
```

```python
param = 'oxygen_mg_per_l'
plt.figure(figsize=(30, 5))
sns.boxplot(x='study_id', y=param, hue='empo_3', data=df_oxygen, linewidth=0.2)
plt.xticks(rotation=45)
plt.tight_layout()
```
